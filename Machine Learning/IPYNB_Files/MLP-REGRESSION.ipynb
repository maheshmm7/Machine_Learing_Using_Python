### CREATING DUMMY VARIABLES

```python
import pandas as pd

# Assuming your DataFrame is already loaded and named 'df'
# If not, load your data into a DataFrame first

# Create dummy variables for 'Airport' column
dummies = pd.get_dummies(df['airport'], prefix='airport')

# Concatenate the dummy variables with the original DataFrame
df = pd.concat([df, dummies], axis=1)

# Drop the original 'Airport' column since we now have dummy variables
df.drop('airport', axis=1, inplace=True)

# Now, you will have separate columns 'Airport_NO' and 'Airport_YES'
# containing 0s and 1s based on the original 'Airport' column values
```

```python
df.head()
```

```python
df.head()
```

```python
waterbody_dummies = pd.get_dummies(df['waterbody'])

waterbody_dummies.rename(columns={
    'Lake and River': 'waterbody_Lake_River',
    'Lake': 'waterbody_Lake',
    'None': 'waterbody_None',
    'River': 'waterbody_River'
}, inplace=True)

df = pd.concat([df, waterbody_dummies], axis=1)

df.drop('waterbody', axis=1, inplace=True)
```

```python
df.head()
```

```python
import pandas as pd



# Convert the 'waterbody' dummy columns to float64 data type
df['waterbody_Lake_River'] = pd.to_numeric(df['waterbody_Lake_River'])
df['waterbody_None'] = pd.to_numeric(df['waterbody_None'])
df['waterbody_River'] = pd.to_numeric(df['waterbody_River'])

# Now all the dummy variables for 'airport' and 'waterbody' are in float64 data type
# Convert 'n_hos_beds' to numeric (float64) data type
df['n_hos_beds'] = pd.to_numeric(df['n_hos_beds'], errors='coerce')
mean_n_hos_beds = df['n_hos_beds'].mean()
df['n_hos_beds'].fillna(mean_n_hos_beds, inplace=True)
```
